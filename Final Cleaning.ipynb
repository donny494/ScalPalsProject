```python
#dataframe of current NBA players
dfcur=pd.read_csv("Data/curdf.csv")
print dfcur.shape
dfcur.head()
```

```python
#remove duplicate/unnecessary columns
dfcur=dfcur.drop(['Player.1','Player.1_x', 'Player.1_y', 'Lg'], 1)
print dfcur.shape
dfcur.head()
```

```python
#checks and merges if columns from both dataframes match
if df.columns.all() == dfcur.columns.all():
    dfnew=pd.concat([df, dfcur],ignore_index=True)
    print dfnew.shape
else:
    print "mismatched columns"
```

```python
dfnew.head()
```

```python
#drops players who did not play in a season
noplay = dfnew[dfnew['Tm'].str.contains("Did").values == True].index.tolist()
for a in noplay:
    try: 
        player = dfnew['Player'][a]
    except:
        pass
    mask=(dfnew['Player'] == player).values
    dfnew = dfnew[~mask]
dfnew[:50]
```

```python
#resets index 
dfnew.reset_index(drop=True, inplace=True)
print dfnew.shape
```

```python
#strings to float
tofloat = ['2P','2P%','2PA','3P','3P%','3PA','AST','Age','BLK','DRB','FG','FG%','FGA','FT','FT%','FTA','G','GS','MP','ORB','PF','PTS','STL','TOV','TRB','eFG%','Salary']
for x in tofloat:
    dfnew[x] = dfnew[x].astype(float)
```

```python
#removes rows contributing to TOT values
#We're looking at total seasons, not portions of seasons
tot = dfnew[dfnew['Tm'].values == "TOT"].index.tolist()
for a in tot:
    for x in range(1,16):
        if a+1 in tot:
            tot.remove(a+1)
        if a+2 in tot:
            tot.remove(a+2)
            if a+3 in tot:
                print a, dfnew['Player'][a], dfnew['Season'][a],"error"
        if a+x<len(dfnew):
            if dfnew['Player'][a+x]==dfnew['Player'][a] and dfnew['Season'][a+x]==dfnew['Season'][a]:
                dfnew.drop([a+x], inplace = True)
```

```python
#resets index
dfnew.reset_index(drop=True, inplace=True)
dfnew.head()
```

```python
#replaces NaN values with 0
nans=['2P%','3P%','FG%','FT%','eFG%']
for x in nans:
    dfnew[x].fillna(0, inplace=True)
dfnew.head()
```

Here we created a function that adjusts for inflation so that we can compare salaries.  We then use this function, in conjunction with inflation data we collected, to replace salaries in the dataframe with the adjusted counterparts.

```python
# We got our data from: http://www.bls.gov/data/inflation_calculator.htm
def inflator(salary, year, rate_dict):
    rate = rate_dict[year]
    return rate*salary + salary

InflationRates = pd.read_csv('Data/Inflation Rates.csv')
Years = InflationRates.Year.values
Rates = InflationRates.Rate.values

InflationDict = dict()
for i in range(len(Years)):
    InflationDict[Years[i]] = Rates[i]

salaries = dfnew.Salary.values
years = dfnew.Season.values
new_vals = []
for i in range(len(dfnew.Salary.values)):
    new_vals.append(inflator(salaries[i],int(years[i][:4]),InflationDict))

dfnew['Salary'] = new_vals
#dfnew.to_csv('IncomeStatsDf.csv',index = False,encode='utf-8')
```
